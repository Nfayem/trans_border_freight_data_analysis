##### 2.2.1.1. Base Code for Concatenation

```python
# Define base directory containing all files
base_dir = "../data"

# Function to read and concatenate CSV files
def concat_files(file_list, folder_path):
    dfs = []
    
    for file in file_list:
        file_path = os.path.join(folder_path, file)
        
        if os.path.exists(file_path):
            print(f"Loading: {file_path}")
            df = pd.read_csv(file_path)
            dfs.append(df)
        else:
            print(f"Warning: {file_path} not found!")
    
    return pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()
```

##### 2.2.1.2. 2020 - 2022 Concatenation

```python
# Processing 2020-2022
years_2020_2022 = {
    "2020": ["dot1_ytd_0920.csv", "dot2_ytd_0920.csv", "dot3_ytd_0920.csv"],
    "2021": ["dot1_ytd_1221.csv", "dot2_ytd_1221.csv", "dot3_ytd_1221.csv"],
    "2022": ["dot1_ytd_1222.csv", "dot2_ytd_1222.csv", "dot3_ytd_1222.csv"],
}

final_years = []

for year, files in years_2020_2022.items():
    folder_path = os.path.join(base_dir, str(year))
    print(f"\nProcessing {year} (Files in `{folder_path}`)...\n")
    final_years.append(concat_files(files, folder_path))

# Master file up to 2022
final_df = pd.concat(final_years, ignore_index=True)

# Display results
final_df
```

##### 2.2.1.3. 2023 Concatenation

```python
# Processing 2023
years_2023 = {
    "2023": {
        "dot_1": ["dot1_ytd_0823.csv", "dot1_0923.csv", "dot1_1023.csv", "dot1_1123.csv", "dot1_1223.csv"],
        "dot_2": ["dot2_ytd_0823.csv", "dot2_0923.csv", "dot2_1023.csv", "dot2_1123.csv", "dot2_1223.csv"],
        "dot_3": ["dot3_ytd_0823.csv", "dot3_0923.csv", "dot3_1023.csv", "dot3_1123.csv", "dot3_1223.csv"],
    }}

for year, subfolders in years_2023.items():
    print(f"\nProcessing {year} (Files in `{base_dir}/{year}/dot_x/`)...")
    subfolder_dfs = []
    
    for subfolder, files in subfolders.items():
        folder_path = os.path.join(base_dir, str(year), subfolder)
        print(f"\nProcessing `{subfolder}`...\n")
        subfolder_df = concat_files(files, folder_path)
        subfolder_dfs.append(subfolder_df)

    # Combine all three `dot_x` datasets into one for the year
    if subfolder_dfs:
        final_year = pd.concat(subfolder_dfs, ignore_index=True)
        final_years.append(final_year)

# Master file up to 2023
final_df = pd.concat(final_years, ignore_index=True)

# Display results
final_df
```

```python
# Function to plot categorical trade information
def plot_cat_features(data, features, title_fontsize=16, label_fontsize=14, tick_labelsize=12, palette='viridis'):
    num_features = len(features)  
    n_rows = num_features

    # Create a figure with the required number of rows
    fig, axes = plt.subplots(n_rows, 1, figsize=(13, 14 * n_rows))

    # Ensure axes is always a list (needed when only one feature is given)
    if n_rows == 1:
        axes = [axes]

    # Loop through each feature to plot
    for idx, feature in enumerate(features):
        # Calculate value counts and sort in descending order
        value_counts = data[feature].value_counts(normalize=True).sort_values(ascending=False)
        sorted_categories = value_counts.index
        total = len(data)

        # Determine whether to use a horizontal or vertical bar plot
        if len(sorted_categories) > 20:
            sns.countplot(y=data[feature], palette=palette, order=sorted_categories, ax=axes[idx])

            # Add percentage labels for horizontal bar chart
            for container in axes[idx].containers:
                labels = [f'{h.get_width()/total*100:.2f}%' if h.get_width() > 0 else '' for h in container]
                axes[idx].bar_label(container, labels=labels, padding=3)

            axes[idx].set_ylabel(feature.capitalize(), fontsize=label_fontsize)
            axes[idx].set_xlabel('Count', fontsize=label_fontsize)
        else:
            sns.countplot(x=data[feature], palette=palette, order=sorted_categories, ax=axes[idx])

            # Add percentage labels for vertical bar chart
            for container in axes[idx].containers:
                labels = [f'{h.get_height()/total*100:.2f}%' if h.get_height() > 0 else '' for h in container]
                axes[idx].bar_label(container, labels=labels, padding=3)

            axes[idx].set_xlabel(feature.capitalize(), fontsize=label_fontsize)
            axes[idx].set_ylabel('Count', fontsize=label_fontsize)
            axes[idx].tick_params(axis='x', rotation=45, labelsize=tick_labelsize)

        # Set title
        axes[idx].set_title(f'Distribution of Trade by {feature}', fontsize=title_fontsize)
        axes[idx].tick_params(axis='y', labelsize=tick_labelsize)

    # Display plot
    plt.tight_layout()
    plt.show()

# Plot categorical features
plot_cat_features(
    data=final_eda_df,
    features=cat_features,
    title_fontsize=16,
    label_fontsize=14,
    tick_labelsize=12,
    palette='viridis'
)
```

```python
# Function to plot numerical trade information
def analyse_numeric_features(data, features):

    # Initialize dictionary to store results
    summary_stats = {}

    for feature in features:
        # Drop missing values
        feature_data = data[feature].dropna()

        # Compute summary statistics
        mean_val = feature_data.mean()
        median_val = feature_data.median()

        summary_stats[feature] = {
            "count": feature_data.count(),
            "mean": mean_val,
            "median": median_val,
            "std": feature_data.std(),
            "skewness": feature_data.skew(),
            "kurtosis": feature_data.kurtosis()
        }

        # Create figure with subplots
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        # Histogram with KDE
        axes[0].hist(feature_data, bins=50, alpha=0.5, color="violet", edgecolor="black", density=True, label="Histogram")
        sns.kdeplot(feature_data, color="darkviolet", linewidth=2, label="KDE", ax=axes[0])

        # Add Mean & Median Lines
        axes[0].axvline(mean_val, color="red", linestyle="dashed", linewidth=2, label=f"Mean: {mean_val:.2f}")
        axes[0].axvline(median_val, color="blue", linestyle="dashed", linewidth=2, label=f"Median: {median_val:.2f}")

        # Titles & Labels
        axes[0].legend()
        axes[0].set_title(f"Histogram & KDE of {feature}", fontsize=14, fontweight="bold")
        axes[0].set_xlabel(feature, fontsize=12)
        axes[0].set_ylabel("Density", fontsize=12)

        # Box Plot
        sns.boxplot(y=feature_data, color="skyblue", ax=axes[1])

        # Title & Labels
        axes[1].set_ylabel(feature, fontsize=12)
        axes[1].set_title(f"Box Plot of {feature}", fontsize=14, fontweight="bold")

        # Adjust layout and show plot
        plt.tight_layout()
        plt.show()

    # Convert dictionary to DataFrame for summary statistics
    summary_df = pd.DataFrame(summary_stats).T  

    return summary_df

# Analyse numeric features
uni_num_stats = analyse_numeric_features(final_eda_df, num_features)
uni_num_stats
```

`Comment` 

**Observed Data Discrepancies**  
Officially, all **port codes under the district code `01XX`** should be associated with **Maine (ME)** in the `USASTATE` column. However, a cross-tabulation analysis reveals that some port codes within this district are assigned to multiple U.S. states. This indicates that **U.S. states in the dataset are linked to multiple port codes, including those not officially assigned to their district codes.** The reverse is also observed, where **district codes are linked to multiple states.**  

These discrepancies suggest that **contextual filling of `USASTATE` and `DEPE` using each other must be approached cautiously** to avoid propagating errors. The following factors likely contribute to these inconsistencies:  

1️. **Data Entry Errors** → Manual input mistakes or system errors may have led to incorrect state assignments, particularly for **port code 0101**.  

2️. **Shipments Processed in Different States** → The `USASTATE` column may represent **the consignee or shipper's location** rather than the actual **port of entry or departure (DEPE)**.  

3️. **Transshipment & Multistate Processing** → Shipments often **pass through multiple states** before reaching their final destination, leading to mismatches between `USASTATE` and `DEPE`.  

4️. **Customs or Administrative Handling** → Some shipments are **administratively classified** under a Maine port code but **physically clear customs in another state**.  

5️. **Coding or Data Transformation Issues** → Errors in **data merging, incorrect joins, or transformations** may have resulted in misassignments between `USASTATE` and `DEPE`.  

6️. **Mode of Transport Influence** → Certain **transport modes (e.g., trucking, rail)** allow goods to be logged in **multiple states before reaching their final port**, causing discrepancies.  

**Recommended Imputation Strategy**  
Since external validation sources are unavailable, a **majority-based imputation approach** is recommended:  

- Compute the **most frequent (`mode`) `USASTATE` for each `DEPE`**.  
- **Impute missing `USASTATE` values only if 90%+ of `DEPE` occurrences are in a single state.**  
- If this criterion is not met, apply additional constraints:  
  - **Filter by mode of transport (`DISAGMOT`).**  
  - **Incorporate commodity type, and freight charges as additional filters** before imputing.  

Conversely, **using `USASTATE` to infer `DEPE` can reduce uncertainty**, but in states with **major logistics hubs (e.g., New York, California, Texas), ports handle diverse cargo**, increasing the likelihood of misclassification.

##### 2.3.6.4. What are the primary contributors to carbon emissions in freight transportation, and how can they be reduced?

```python
# Initialize geolocator with caching enabled
geolocator = Nominatim(user_agent="freight_distance_estimator", timeout=10)

# Dictionary to store previously retrieved coordinates
location_cache = {}

# Function to get coordinates
def get_coordinates(location):
    if pd.isna(location):
        return None

    if location in location_cache:
        return location_cache[location]

    try:
        loc = geolocator.geocode(location)
        if loc:
            location_cache[location] = (loc.latitude, loc.longitude)
            return (loc.latitude, loc.longitude)
    except Exception as e:
        print(f"Geocoding error for {location}: {e}")
    
    return None

# Function to estimate distance
def estimate_distance(row):
    origin = get_coordinates(row['USASTATE'])
    destination = get_coordinates(row['MEXSTATE']) or get_coordinates(row['CANPROV'])

    if origin and destination:
        return geodesic(origin, destination).kilometers
    return None 

# Apply function with progress tracking
tqdm.pandas()
final_eda_df['ESTIMATED_DISTANCE'] = final_eda_df.progress_apply(estimate_distance, axis=1)

# Display results
final_eda_df[['USASTATE', 'MEXSTATE', 'CANPROV', 'ESTIMATED_DISTANCE']]
```

The most widely accepted formula for calculating **carbon emissions** in [freight transportation](https://business.edf.org/insights/green-freight-math-how-to-calculate-emissions-for-a-truck-move/) follows this structure:  

![CO<sub>2</sub>_Emissions_formula](../assets/ghg_emission_formula.webp)

This approach considers the total carbon footprint from **all greenhouse gases (GHG)**; including **CO₂ (carbon dioxide), CH₄ (methane), and N₂O (nitrous oxide)**(arranged in ascending order of their **global warming potential (GWP)**).  

**Key Components of the Emission Calculation**  
1️. **Shipment Weight (`SHIPWT`)**  
   - The dataset contains `SHIPWT` in **kilograms (kg)**.  
   - It will be **converted to tonnes** for standard emissions calculations (1 tonne = 1000 kg).  

2️. **Transport Distance (`ESTIMATED_DISTANCE`)**  
   - This column contains the **pre-estimated transport distance** for each trade.  

3️. **Emission Factors (kgCO₂e/tonne.km)**  
   - Derived from the **UK Government's Greenhouse Gas Reporting Conversion Factors (2024)**, specifically from the **Freighting Goods sheet** in the dataset [Greenhouse Gas Reporting: Conversion Factors 2024](https://assets.publishing.service.gov.uk/media/6722566a3758e4604742aa1e/ghg-conversion-factors-2024-condensed_set__for_most_users__v1_1.xlsx).  
   - These values provide **mode-specific emission intensities** based on real-world freight transport data.

![CO<sub>2</sub>_Emissions_for_Freighting_Goods](../assets/carbon_emission_factor.png)

**Emission Factors for Different Freight Transport Modes**  
| **Transport Mode** | **Emission Factor (kgCO₂e/tonne.km)** |
|--------------------|----------------------------------|
| **Vessel (Sea Cargo & Tankers - Avg.)** | 0.02544 |
| **Air (Freight Flights - Avg)** | 1.69775 |
| **Truck (Avg. Laden Heavy Goods Vehicles - Avg.)** | 0.21550 |
| **Rail (Standard Rail Transport)** | 0.02779 |

**Modes Excluded from CO₂ Emissions Calculation**  
Some modes of transport in the dataset are **not associated with direct CO₂ emissions** and will be **excluded** from the calculation:  
- **Mail (U.S. Postal Service)** → Primarily a logistics system, actual transport depends on secondary carriers.  
- **Pipeline** → Used for transporting **liquids or gases**, but does not fit conventional freight emission models.  
- **Other** → A generic category that lacks a clearly defined transport mode.  
- **Foreign Trade Zones (FTZs)** → Designated **customs areas**, not an actual transport mode.  

```python
# # Save the cleaned data to a CSV file
# output_file = '../data/BTS_trade_data.csv'
# final_eda_df.to_csv(output_file, index=False)
```

```python
# Enable tqdm progress tracking
tqdm.pandas()

# Define Emission Factors (kgCO₂e/tonne.km)
emission_factors = {
    "Vessel": 0.02544,  
    "Air": 1.69775,      
    "Truck": 0.21550,    
    "Rail": 0.02779      
}

# Function to estimate CO₂ emissions
def calculate_co2_emissions(row):
    mode = row["DISAGMOT"]
    weight_tonnes = row["SHIPWT"] / 1000
    distance_km = row["ESTIMATED_DISTANCE"]  

    # Ensure valid transport mode and distance
    if mode in emission_factors and pd.notna(distance_km):
        emission_factor = emission_factors[mode]
        return weight_tonnes * distance_km * emission_factor
    
    return 0

# Apply function to dataset with progress tracking
final_eda_df["CO2_EMISSIONS"] = final_eda_df.progress_apply(calculate_co2_emissions, axis=1)

# Display results
final_eda_df[["DISAGMOT", "SHIPWT", "ESTIMATED_DISTANCE", "CO2_EMISSIONS"]]
```

#### 2.3.7. Hypothesis
Validate hypotheses through statistical methods (e.g., correlation analysis, Chi-square tests) to assess the impact of transportation mode selection, risk management strategies, and sustainability measures on freight transportation efficiency, environmental impact, and resilience. 

**A. Null Hypothesis (H<sub>0</sub>):**  
Transportation mode selection, risk management strategies, and sustainability measures have no statistically significant effect on freight transportation efficiency, environmental impact, or resilience.

**B. Alternative Hypothesis (H<sub>1</sub>):**  
Optimized transportation mode selection, risk management strategies, and sustainability measures significantly improve freight transportation efficiency, reduce environmental impact, and enhance resilience.

Based on Carbon Intensity & Efficiency: Arrange transport modes from highest to lowest emissions per unit freight moved, ensuring a realistic impact on efficiency and cost. This allows a clear negative correlation with emissions and costs (since higher-ranked modes are more efficient).  
- `Truck`: 1 
- `Vessel`: 2 
- `Rail`: 3 
- `Air`: 4 
- `Other, Mail (U.S. Postal Service), Foreign Trade Zones (FTZs), Pipeline`: 5

```python
# Define custom ordering
transport_order = {
    'Truck': 1, 'Vessel': 2, 'Rail': 3, 'Air': 4, 'Other': 5, 'Mail (U.S. Postal Service)': 5, 'Foreign Trade Zones (FTZs)': 5, 'Pipeline': 5
}

# Convert categorical to ordered numerical values
final_eda_df['DISAGMOT_NUM'] = final_eda_df['DISAGMOT'].map(transport_order)

# Compute correlation matrix with ordered DISAGMOT
corr_features = ['DISAGMOT_NUM', 'FREIGHT_CHARGES', 'SHIPWT', 'VALUE', 'CO2_EMISSIONS']
corr_matrix = final_eda_df[corr_features].corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Analysis: Transport Mode vs Efficiency & Cost')
plt.show()
```

```python
# Create a contingency table between Mode of Transport (DISAGMOT) and DF (Domestic/Foreign)
contingency_table = pd.crosstab(final_eda_df['DISAGMOT'], final_eda_df['DF'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Display results
print(f'Chi-Square Statistic: {chi2}')
print(f'p-value: {p}')

# Interpretation
if p < 0.05:
    print("Reject the Null Hypothesis: Transportation mode selection significantly impacts sustainability.")
else:
    print("Fail to Reject the Null Hypothesis: No significant impact found.")

# Count occurrences
mode_counts = pd.crosstab(final_eda_df['DISAGMOT'], final_eda_df['DF'])

# Plot stacked bar chart
mode_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 8))
plt.title('Transport Mode Usage by Domestic/Foreign Trade')
plt.xlabel('Mode of Transport')
plt.ylabel('Count')
plt.legend(title="DF (Domestic/Foreign)")
plt.xticks(rotation=45)
plt.show()
```

**Kolmogorov-Smirnov (KS) test**

The KS test checks whether two samples come from the same distribution.  
- **Null Hypothesis (H₀):** The sampled data follows the same distribution as the full dataset.  
- **Alternative Hypothesis (H₁):** The sampled data is significantly different from the full dataset.  

If **p-value > 0.05**, we **fail to reject H₀** → The sample preserves the original distribution.  
If **p-value < 0.05**, we **reject H₀** → The sample does not preserve the original distribution.

```python
# Define sampling percentages
sample_sizes = [0.0001, 0.0002, 0.0003, 0.0005, 0.05, 0.1, 0.2]
features = ["SHIPWT"]

# Store KS test results
ks_results = {}

for frac in sample_sizes:
    sampled_df = final_eda_df.sample(frac=frac, random_state=42)
    
    ks_results[frac] = {}
    
    for feature in features:
        stat, p_value = ks_2samp(final_eda_df[feature], sampled_df[feature])
        ks_results[frac][feature] = {"KS Statistic": stat, "p-value": p_value}

# Convert results to DataFrame
ks_results_df = pd.DataFrame.from_dict(
    {(i, j): ks_results[i][j] for i in ks_results.keys() for j in ks_results[i].keys()},
    orient='index'
).reset_index()

ks_results_df.columns = ["Sample Size", "Feature", "KS Statistic", "p-value"]
ks_results_df
```

`Comment`

**🚀 Key Observations from the KS Test Results:**
| **Sample %** | **KS Statistic** | **p-value** | **Interpretation** |
|-------------|---------------|-----------|----------------|

**Best Sampling Choices for Speed & Accuracy**  

- 0.02% (~1,300 rows) or 0.03% (~1,950 rows) → Faster but may miss rare patterns.
  
- 5%+ (~65,000+ rows) → Highly accurate but too slow for quick analysis.
  
- Optimal Trade-off is 0.05% (~3,250 rows) 
  - Strong representativeness (p > 0.05), minimal deviation from full dataset.  
  - More computationally efficient than 5% or higher samples.
